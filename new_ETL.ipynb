## creamos una función para convertir las cadenas en un diccionario real

```python
import ast
def cadena_diccionario(value):
    if pd.isna(value):
        return None
    else:
        return ast.literal_eval(value)
```

# Desanidamos la columnas 'belong_to_collection'

```python
# Convertimos los elementos de la columna a un diccionario real
data["belongs_to_collection"] = data["belongs_to_collection"].apply(cadena_diccionario)
```

```python
data.belongs_to_collection
```

```python
# Función para extraer el valor de la clave "name" de los diccionario en la columna "collection"
def extract_name(record):
    if record is not None:
        if isinstance(record, dict):
            return record.get('name')
    else:
        return None

# Aplicar la función a la columna "collection" y crear una nueva columna "name_to_collection"
data['name_to_collection'] = data['belongs_to_collection'].apply(extract_name)
#mostramos dataset
data
```

# Desanidamos la columna 'genres'

```python
# Convertimos a diccionario real los valores que estan en cadena de string en dicha columna
data['genres'] = data['genres'].apply(cadena_diccionario)

# Función para extraer los valores de la clave "name" del diccionario
def extract_name(record):
    return [item['name'] for item in record]

# Aplicar la función a la columna 'genres' y expandir los resultados en una nueva columna 'name_genres'
data['name_genres'] = data['genres'].apply(lambda x: extract_name(x))
#mostrar dataFrama
data.head(5)
```

# Funcion que vamos a usar para desanidar las columnas restantes

```python
def desanida_name(columna, lugar_name_int): # en lugar_name_int hay que colocar en que posicion del diccionario esta "name"
    global data
    # Funcion para convertir de diccionario a string
    import ast
    def parse_dict_string(value):
        if pd.isna(value):
            return None
        else:
            return ast.literal_eval(value)
        
    # Aplicamos la funcion para pasar de diccionario a string y desanidamos
    data[columna] = data[columna].apply(parse_dict_string)
    df_desanidado = pd.json_normalize(data[columna]).fillna("")
    #return df_desanidado.head()
    
    # Obtener la cantidad de columnas en el DataFrame
    cantidad_columnas = len(df_desanidado.columns)

    # Creamos un nuevo dataframe para concatenarlo luego de desanidar completamente el dataframe anterior (df_desanidado)
    concatenar_data = pd.DataFrame()

    # Iterar sobre las columnas del DataFrame
    for i in range(cantidad_columnas):
        df_desanidado_2 = pd.json_normalize(df_desanidado[i]).fillna("")
        # Concatenar los dataframes
        concatenar_data = pd.concat([concatenar_data, df_desanidado_2], axis=1)
    
    # Función para juntar los valores en una lista
    def juntar_valores_en_lista(row):
        lista_valores = []
        for valor in row:
            if valor.strip():  # Verificar si el valor no está vacío después de eliminar espacios en blanco
                lista_valores.append(valor)
        return lista_valores

    # Aplicar la función a las columnas y crear una nueva columna 'Lista'
    concatenar_data['name'] = concatenar_data["name"].apply(juntar_valores_en_lista, axis=1)

    # Selección de columnas por número de columna
    df_seleccionado = concatenar_data.iloc[:, lugar_name_int]
    df_seleccionado = pd.DataFrame(df_seleccionado)

    # Concatenamos df_desanidado["name"] a data
    data = pd.concat([data, df_seleccionado['name']], axis=1)
```

# Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0. Luego creamos la columna 'return' que va a ser la el resultado de la division de los valores de la columna 'revenue' y 'budget' (revenue / budget)

```python
# Los datos de la columna 'budget' son del tipo string y lo pasamos al tipo de dato numerico
data['budget'] = pd.to_numeric(data['budget'], errors='coerce')
data['budget']
```

```python
data['budget']=data['budget'].fillna(0)
data['revenue']=data['revenue'].fillna(0)
data['return'] = data['revenue'] / data['budget']
data['return'] = data['return'].fillna(0)
data['return'] = data['return'].replace([np.inf, -np.inf], 0)
data[['revenue', 'budget', 'return']]
```

## De haber fechas, deberán tener el formato AAAA-mm-dd, además deberán crear la columna release_year donde extraerán el año de la fecha de estreno.

```python
# Cambiamos fechas a AAAA/MM/DD
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')
data['release_date'] = pd.to_datetime(data['release_date']).dt.strftime('%Y-%m-%d')

# Eliminamos los registros nulos
data = data.dropna(subset=['release_date'])
data['release_date'].isnull().sum()
data['release_date'] = pd.to_datetime(data['release_date'])
data['release_year'] = data['release_date'].dt.year
data[['release_date', 'release_year']].tail(10)
```

# Ahora realizamos las transformaciones necesarias para el dataset 'credits.csv'

```python
creditos=pd.read_csv('credits.csv')
creditos.head(5)
```

```python
creditos.shape
```

```python
import re
#extraemos solo los nombres de las coleccion
patron = r"'name': '([^']*)"
#extreemos nom coleccion
creditos['director'] = creditos['crew'].str.extract(patron, flags= re.IGNORECASE )
#columna valores individuales
creditos['name_director']=creditos['director'].str.split(', ')
#se duplican los registros
creditos = creditos.explode('name_director').reset_index(drop= True)
#se elimina la columna 'coleccion_name
creditos.drop('director', axis=1, inplace= True) 
#mostramos
creditos.head(5)
```
